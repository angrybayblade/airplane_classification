```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2

from sklearn.metrics import confusion_matrix
from tqdm.notebook import tqdm

%matplotlib inline
```

```python
import tensorflow as tf

from tensorflow import keras 
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import LearningRateScheduler,ModelCheckpoint,Callback
from tensorflow.keras.regularizers import l1,l2,l1_l2
```

```python
train = pd.read_csv("./train.csv")
val = pd.read_csv("./validation.csv")
```

```python
train_labels = pd.get_dummies(train.family)
val_labels = pd.get_dummies(val.family)

train_classes = train_labels.columns
val_classes = val_labels.columns

train_labels = train_labels.values
val_labels = val_labels.values
```

```python
images_train = np.array([
    cv2.resize(
        cv2.cvtColor(
            cv2.imread(i),
            cv2.COLOR_BGR2RGB
        ),
        (96,64),
        interpolation=cv2.INTER_CUBIC
    ) / 255
    
    for 
        i
    in
        tqdm(train.image.values)
]).astype(np.float32)
```

```python
images_val = np.array([
    cv2.resize(
        cv2.cvtColor(
            cv2.imread(i),
            cv2.COLOR_BGR2RGB
        ),
        (96,64),
        interpolation=cv2.INTER_CUBIC
    ) / 255
    
    for 
        i
    in
        tqdm(val.image.values)
]).astype(np.float32)
```

### Modeling

```python
def conv_2d(x,filters,ksize,strides=1,activation=ReLU,drop=False):
    x = Conv2D(
            filters,
            ksize,
            strides=strides,
            padding="same",
        )(x)
    x = BatchNormalization()(x)
    x = activation()(x)
    
    return x

def depthwise_sep_conv_2d(x,ksize,strides=1,activation=ReLU,drop=False):
    x = DepthwiseConv2D(
            kernel_size=ksize,
            strides=strides,
            padding="same",
        )(x)
    x = BatchNormalization()(x)
    x = activation()(x)
    if drop:
        x = SpatialDropout2D(0.15)(x)
    return x
    

    
    
_in = Input((64,96,3))
a = conv_2d(_in,32,7)

b = conv_2d(a,16,1)
c = depthwise_sep_conv_2d(b,3,strides=2)

d = conv_2d(c,32,1)
e = depthwise_sep_conv_2d(d,3,strides=2)

f = conv_2d(e,64,1)
g = depthwise_sep_conv_2d(f,3,strides=2)

h = conv_2d(g,128,1)
i = depthwise_sep_conv_2d(h,3,strides=2)

j = conv_2d(i,256,1)
k = depthwise_sep_conv_2d(j,3)


flat = GlobalAvgPool2D()(k)
fc = Dense(512)(flat)
fc = BatchNormalization()(fc)
fc = ReLU()(fc)

out = Dense(70)(fc)
out = BatchNormalization()(out)
out = Activation("softmax")(out)


model = keras.Model(_in,out)
```

```python
model.summary()
```

```python
keras.utils.plot_model(model)
```

```python
def flow(x,y,batch_size):
    br = len(x)%batch_size
    index = np.arange(0,len(x))
    while True:
        index = index.reshape(-1)
        np.random.shuffle(index)
        if br:
            index_ = index[:-br].reshape(-1,batch_size)
            for batch in index_:
                yield x[batch],y[batch]
            yield x[index[-br:]],y[index[-br:]]
        else:
            index_ = index.reshape(-1,batch_size)
            for batch in index_:
                yield x[batch],y[batch]
```

```python
def schedule(epoch):
    if epoch < 60:
        return .009
    else:
        return .0009
```

```python
image_data_gen = keras.preprocessing.image.ImageDataGenerator(
    rotation_range=20,
    vertical_flip=True
)
```

```python
EPOCHS = 100
BATCH_SIZE = 16
STEPS_PER_EPOCHS = np.ceil(len(images_train)/BATCH_SIZE)

F_TRAIN = image_data_gen.flow(x=images_train,y=train_labels,batch_size=BATCH_SIZE)#flow(images_train,train_labels,BATCH_SIZE)
F_VAL = flow(images_val,val_labels,BATCH_SIZE)
```

```python
opt = keras.optimizers.Adam(learning_rate=0.009,beta_1=0.99,beta_2=0.997,epsilon=1e-8)
loss = keras.losses.CategoricalCrossentropy()
```

```python
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
lrs = LearningRateScheduler(schedule=schedule,verbose=1)
ckp = ModelCheckpoint(filepath="./ckp/CNN_0_ckps",save_weights_only=True)
```

```python
history = model.fit_generator(
            generator=F_TRAIN,
            steps_per_epoch=STEPS_PER_EPOCHS,
            epochs=EPOCHS,
            validation_data=F_VAL,
            validation_steps=STEPS_PER_EPOCHS,
            callbacks=[lrs,ckp]
    )
```

```python
hist = pd.DataFrame(history.history)
```

```python
plt.plot(hist.index,hist.val_loss,label="Validation Loss")
plt.plot(hist.index,hist.loss,label="Training Loss")
plt.legend()
```

```python
plt.plot(hist.index,hist.val_accuracy,label="Validation Accuracy")
plt.plot(hist.index,hist.accuracy,label="Training Accuracy")
plt.legend()
```

### Validation

```python
### training data

train_pred = model.predict(images_train,verbose=1,batch_size=32)
y_pred = train_pred.argmax(-1)
y_true = train_labels.argmax(-1)
```

```python
conf_mat = confusion_matrix(y_true,y_pred)
```

```python
fig, ax = plt.subplots(figsize=(14,14))
im = ax.imshow(conf_mat)

ax.set_xticks(np.arange(len(train_classes)))
ax.set_yticks(np.arange(len(train_classes)))

ax.set_xticklabels(train_classes)
ax.set_yticklabels(train_classes)

plt.setp(ax.get_xticklabels(), rotation=90, ha="right",
         rotation_mode="anchor")

for i in range(len(train_classes)):
    for j in range(len(train_classes)):
        if i == j:
            text = ax.text(j, i, conf_mat[i, j],
                           ha="center", va="center", color="w")
fig.tight_layout()
plt.show()
```

```python
classwise_acc = (conf_mat.astype('float') / conf_mat.sum(axis=1)[:, np.newaxis]).diagonal()
```

```python
fig = plt.figure(figsize=(8,20))
plt.barh(width=classwise_acc,y=list(train_classes))
plt.xlabel("Accuracy")
plt.ylabel("Class")
plt.title("Classwise Accuracy")
```
