```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
train = pd.read_csv("./train.csv")
val = pd.read_csv("./validation.csv")
```

```python
plt.figure(figsize=(12,36))
train.family.value_counts().plot.barh()
```

```python
plt.figure(figsize=(12,36))
val.family.value_counts().plot.barh()
```

```python
train_group = train.groupby("family")
train_groups = list(train_group.groups.keys())
```

```python
group = train_groups[4]
group = train_group.get_group(group)
images = group.image.values[:9]#np.random.choice(group.image.values,size=(9,))

fig,axes = plt.subplots(nrows=3,ncols=3,figsize=(12,10)) 
for img,axs in zip(images,axes.reshape(-1)):
    img = cv2.resize(cv2.cvtColor(cv2.imread(img),cv2.COLOR_BGR2RGB),(96,64))
    axs.imshow(img)

print ("Family : ",group.iloc[0,2])
```
